```python
## upravená funkcia -konzultovana s Gemini - definovanie funkcie 
def vypocet_cielovej_sumy(jednorazovy_vklad, pravidelny_vklad, urokova_sadzba_rocne_perc, pocet_rokov):
    ## Vypočíta buducu hodnotu (Future Value - FV) sporenia kombinujúceho jednorazový a pravidelný
    ## mesačný vklad.(Používa float, úročenie je mesačné).
        
    # 1. Konverzia vstupov
    urokova_sadzba_decimal = urokova_sadzba_rocne_perc / 100 # Prevod percenta na desatinné číslo
    urokova_sadzba_mesacne = urokova_sadzba_decimal / 12
    pocet_obdobi = pocet_rokov * 12
    
    # Prehľadnejšie premenné- skratky
    r = urokova_sadzba_mesacne
    n = pocet_obdobi
    
    # 2. Výpočet FV jednorazového vkladu (FV = PV * (1 + r)^n)
    # Zložené úročenie
    fv_jednorazova = jednorazovy_vklad * ((1 + r) ** n)
    
    # 3. Výpočet FV pravidelných vkladov (Anuita Future Value)
    # FV = P * [((1 + r)^n - 1) / r]
    if r == 0:
        # Ochrana proti deleniu nulou (ak je úrok 0%)- toto je dôležité!
        fv_pravidelne = pravidelny_vklad * n
    else:
        fv_pravidelne = pravidelny_vklad * (((1 + r) ** n - 1) / r)
        
    # 4. Celková cieľová suma a zaokrúhlenie na 2 desatinné miesta
    cielova_suma = fv_jednorazova + fv_pravidelne
    
    return round(cielova_suma, 2)
```

```python
##volanie funkcie 
print(vypocet_cielovej_sumy(1000,50,5,10))
```

```python
def simuluj_vyvoj_investicie(jednorazovy_vklad, pravidelny_vklad, urokova_sadzba_rocne_perc, pocet_rokov):
    ##Simuluje vývoj investície mesiac po mesiaci a vracia zoznam hodnôt.

   ## Args:
   ##    jednorazovy_vklad (float): Počiatočný kapitál.
   ##    pravidelny_vklad (float): Mesačný vklad.
   ##    urokova_sadzba_rocne_perc (float): Ročná úroková sadzba v percentách (napr. 5).
   ##    pocet_rokov (int): Doba sporenia v rokoch.

   ## Returns:
   ##   list: Zoznam hodnôt investície na konci každého mesiaca.

    ## Konverzia vstupov 
    urokova_sadzba_decimal = urokova_sadzba_rocne_perc / 100
    urokova_sadzba_mesacne = urokova_sadzba_decimal / 12
    pocet_mesiacov = pocet_rokov * 12
    
    ## vytvorenie prázdneho zoznamu na ukladanie hodnôt
    historia_hodnot = []
    aktualna_hodnota = jednorazovy_vklad # Začíname s jednorazovým vkladom

    for mesiac in range(1, pocet_mesiacov + 1): 
        # Pripočíta úroky k aktuálnej hodnote
        aktualna_hodnota *= (1 + urokova_sadzba_mesacne)
        
        # Pripočíta pravidelný mesačný vklad (ak je prvý mesiac, už je tam jednorazový vklad)
        aktualna_hodnota += pravidelny_vklad
        
        # Uložíme hodnotu zaokrúhlenú na centy
        historia_hodnot.append(round(aktualna_hodnota, 2))
        
    return historia_hodnot
```

```python
## volanie funkcie - vytvorí mi zoznam hodnôt
print(simuluj_vyvoj_investicie(1000, 100, 8, 20))
```

```python
import plotly.graph_objects as go
import pandas as pd
from datetime import date, timedelta
import plotly.io as pio
# --- VSTUPNÉ HODNOTY OD POUŽÍVATEĽA (príklady) ---
jednorazovy_vklad_input = 1000   # Napr. €
pravidelny_vklad_input = 100     # Napr. € mesačne
urokova_sadzba_input = 8         # V percentách, napr. 5 pre 5%
pocet_rokov_input = 20           # V rokoch

# --- 1. Simulácia vývoja investície ---
hodnoty_investicie = simuluj_vyvoj_investicie(
    jednorazovy_vklad_input,
    pravidelny_vklad_input,
    urokova_sadzba_input,
    pocet_rokov_input
)

# --- 2. Príprava dát pre graf ---
# Vytvoríme časovú os pre graf (mesiace)
datum_start = date.today()
mesiace = [datum_start + timedelta(days=30*i) for i in range(len(hodnoty_investicie))]

# Pandas DataFrame pre jednoduchšiu prácu s dátami
df = pd.DataFrame({
    'Dátum': mesiace,
    'Hodnota_Investicie': hodnoty_investicie,
})

# --- 3. Vytvorenie grafu pomocou Plotly ---
fig = go.Figure()


fig.add_trace(go.Scatter(
    x=df['Dátum'],
    y=df['Hodnota_Investicie'],
    mode='lines',
    name='Vývoj Investície',
    line=dict(color='royalblue', width=2),
    ))

# --- 4. Pridanie detailov a titulkov k grafu ---
fig.update_layout(
    title=f'Vývoj Investície v Čase ({pocet_rokov_input} rokov)',
    xaxis_title='Čas',
    yaxis_title='Hodnota Investície (€)',
    hovermode='x unified', # Pre lepšiu interaktivitu
    template='plotly_white', # Pre čistý vzhľad
    )

# Zobrazí graf
fig.show()
```
